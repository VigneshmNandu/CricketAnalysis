```python
df1= pd.DataFrame(df.groupby('striker')['runs_off_bat'].sum()).reset_index()
```

```python
df1.head(20)
```

```python
df2= pd.DataFrame(df.groupby('striker')['ball'].count()).reset_index()
```

```python
df2.head(10)
```

```python
df3 = df1.merge(df2, on = 'striker', how = 'left')
```

```python
df3.head(10)
```

```python
df3= df3[df3.runs_off_bat >= 400]
```

```python
df3.head(10)
```

```python
df3['strike_rate']= df3['runs_off_bat']/df3['ball']*100
```

```python
df3.reset_index()
```

```python
df3= df3[df3.strike_rate >= 150]
```

```python
df3.groupby(['striker'])['strike_rate'].sum().sort_values(ascending = False).plot(kind ='barh')
plt.xlabel('strike_rate')
plt.show()
```

```python
len(df3)
```

```python
df3.reset_index(inplace=True, drop=True)
df3
```

```python
df3.groupby(['striker'])['runs_off_bat'].sum().sort_values(ascending = False).plot(kind ='barh')
plt.xlabel('Runs Scored')
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(18,9))
plt.scatter(df3.strike_rate, df3.runs_off_bat)

for i in range(len(df3)):
     plt.text(df3['strike_rate'][i] + 1, df3['runs_off_bat'][i] - 1, df3['striker'][i] )
       
plt.title('Batsmen with more than 400 runs scored with strike rate more than 150 (Finishers) from IPL 2022-2024', fontsize=26 )
plt.xlabel('Strike rate', fontsize=16)
plt.ylabel('Runs scored', fontsize=16)
plt.axvline(145, ls='--', color='black')
plt.axhline(400, ls='--', color='black')
plt.show()
```

```python
latestIPL_df2 =  pd.DataFrame(latestIPL_df.groupby('striker')['ball'].count()).reset_index()
```

```python
latestIPL_df2.head(10)
```

```python
latestIPL_df3 = latestIPL_df1.merge(latestIPL_df2, on = 'striker', how = 'left')
latestIPL_df3.head()
```

```python

latestIPL_df3 = latestIPL_df3[latestIPL_df3.runs_off_bat >= 100]
latestIPL_df3 = latestIPL_df3[latestIPL_df3.strike_rate >= 160]
latestIPL_df3.reset_index(inplace=True, drop=True)
latestIPL_df3
```

```python
latestIPL_df3.groupby(['striker'])['runs_off_bat'].sum().sort_values(ascending = True).plot(kind ='barh')
plt.xlabel('Runs scored')
plt.show()
```

```python
latestIPL_df3.groupby(['striker'])['strike_rate'].sum().sort_values(ascending = True).plot(kind ='barh')
plt.xlabel('strike_rate')
plt.show()
```

```python
latestIPL_df3.groupby(['striker'])['ball'].sum().sort_values(ascending = False).plot(kind ='barh')
plt.xlabel('No of balls played')
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(18,9))
plt.scatter(latestIPL_df3.strike_rate, latestIPL_df3.runs_off_bat)

for i in range(len(latestIPL_df3)):
     plt.text(latestIPL_df3['strike_rate'][i] + 1, latestIPL_df3['runs_off_bat'][i] - 1, latestIPL_df3['striker'][i] )
        
plt.title('Batsmen with more than 100 runs scored with Strike rate more that 160(Finishers) in this year IPL (2024)', fontsize=26 )
plt.xlabel('Strike rate', fontsize=16)
plt.ylabel('Runs scored', fontsize=16)
plt.axvline(160, ls='--', color='black')
plt.axhline(100, ls='--', color='black')
plt.show()
```
